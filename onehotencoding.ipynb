```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
data = {
    'City': ['Pune', 'Mumbai', 'Pune', 'Delhi'],
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'Age': [23, 25, 22, 30]
}

df = pd.DataFrame(data)
print(df)
```

```python
encoder = OneHotEncoder(sparse=False)

encoded_data = encoder.fit_transform(df[['City', 'Gender']])
```

```python
encoded_df = pd.DataFrame(
    encoded_data,
    columns=encoder.get_feature_names_out(['City', 'Gender'])
)

print(encoded_df)
```

```python
final_df = pd.concat([encoded_df, df[['Age']]], axis=1)
print(final_df)
```

# Another WAY

```python
import pandas as pd

data = {
    'City': ['Pune', 'Mumbai', 'Pune', 'Delhi'],
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'Age': [23, 25, 22, 30]
}

df = pd.DataFrame(data)
df
```

```python
#best WAY
cat_cols = df.select_dtypes(include=['object', 'category']).columns

df_dummies = pd.get_dummies(df, columns=cat_cols, drop_first=True)

df_dummies = df_dummies.astype(int)
```
